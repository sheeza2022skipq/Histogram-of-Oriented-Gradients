```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt
import math
import os
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
```

```python
#gradient , magnitude and orientation calculation
def gradient_calculation(im):
    rows,columns = im.shape
    #im = np.float64 (im )
    Sobel_x_im = np.zeros((rows,columns))
    Sobel_y_im = np.zeros((rows,columns))
    magnitude_image = np.zeros((rows,columns))
    orientation_image = np.zeros((rows,columns))
    
    Sobel_x = np.array([[1, 0, -1],
                        [2, 0, -2],
                        [  1, 0, -1]])


    Sobel_y = np.array([[1, 2, 1],
                        [0, 0, 0],
                        [-1, -2, -1]])

    for i in range(1,rows-1):
        for j in range(1,columns-1):
            Sobel_x_im[i,j]=(im[i-1,j-1]*Sobel_x[0][0])+(im[i-1,j]*Sobel_x[0][1])+(im[i-1,j+1]*Sobel_x[0][2])+(im[i,j-1]*Sobel_x[1][0])+(im[i,j]*Sobel_x[1][1])+(im[i,j+1]*Sobel_x[1][2])+(im[i+1,j-1]*Sobel_x[2][0])+(im[i+1,j]*Sobel_x[2][1]) + (im[i+1,j+1]*Sobel_x[2][2])
            Sobel_y_im[i,j]=(im[i-1,j-1]*Sobel_y[0][0])+(im[i-1,j]*Sobel_y[0][1])+(im[i-1,j+1]*Sobel_y[0][2])+(im[i,j-1]*Sobel_y[1][0])+(im[i,j]*Sobel_y[1][1])+(im[i,j+1]*Sobel_y[1][2])+(im[i+1,j-1]*Sobel_y[2][0])+(im[i+1,j]*Sobel_y[2][1]) + (im[i+1,j+1]*Sobel_y[2][2])
   
    """
    plt.figure() 
    plt.imshow(Sobel_x_im,cmap="gray")
    plt.title("Gradient in x-direction")
    plt.figure()
    plt.imshow(Sobel_y_im,cmap="gray")
    plt.title("Gradient in y-direction")
    """
    for i in range(rows):
        for j in range(columns):
            xx = (Sobel_x_im[i,j])**2
            yy = (Sobel_y_im[i,j])**2
            magnitude_image[i,j] = math.sqrt(xx + yy)
            x = (Sobel_x_im[i,j])
            y = (Sobel_y_im[i,j])
            #radians
            #orientation_image[i,j]= np.arctan2(y,x)
            #degrees
            orientation_image[i,j]= np.absolute(np.rad2deg(np.arctan2(y,x)))
                    
    #print(np.min(orientation_image))
    #print(np.max(orientation_image))   
    return magnitude_image, orientation_image
```

```python
def get_bin(angle):
    if angle == 0:
        d1= d2= r1= r2 = flag = 0
        bins = 0
        return bins,flag, d1, d2, r1, r2
    elif angle == 20:
        d1= d2= r1= r2 = flag = 0
        bins = 1
        return bins,flag, d1, d2, r1, r2
    elif angle == 40:
        d1= d2= r1= r2 = flag = 0
        bins = 2
        return bins,flag, d1, d2, r1, r2
    elif angle == 60:
        d1= d2= r1= r2 = flag = 0
        bins = 3
        return bins,flag, d1, d2, r1, r2
    elif angle == 80:
        d1= d2= r1= r2 = flag = 0
        bins = 4
        return bins,flag, d1, d2, r1, r2
    elif angle == 100:
        d1= d2= r1= r2 = flag = 0
        bins = 5
        return bins,flag, d1, d2, r1, r2
    elif angle == 120:
        d1= d2= r1= r2 = flag = 0
        bins = 6
        return bins,flag, d1, d2, r1, r2
    elif angle == 140:
        d1= d2= r1= r2 = flag = 0
        bins = 7
        return bins,flag, d1, d2, r1, r2
    elif angle == 160:
        d1= d2= r1= r2 = flag = 0
        bins = 8
        return bins,flag, d1, d2, r1, r2


    elif 0 < angle < 20:
        d1=np.absolute(angle-0)
        d2=np.absolute(angle-20)
        r1=d1/20
        r2=d2/20
        flag = 1
        bins = 0
        return bins, flag, d1, d2, r1, r2
    elif 20 < angle < 40:
        d1=np.absolute(angle-20)
        d2=np.absolute(angle-40)
        r1=d1/20
        r2=d2/20
        flag = 1
        bins = 1
        return bins, flag, d1, d2, r1, r2
    elif 40 < angle < 60:
        d1=np.absolute(angle-40)
        d2=np.absolute(angle-60)
        r1=d1/20
        r2=d2/20
        flag = 1
        bins = 2
        return bins, flag, d1, d2, r1, r2
    elif 60 < angle < 80:
        d1=np.absolute(angle-60)
        d2=np.absolute(angle-80)
        r1=d1/20
        r2=d2/20
        flag = 1
        bins = 3
        return bins, flag, d1, d2, r1, r2
    elif 80 < angle < 100:
        d1=np.absolute(angle-80)
        d2=np.absolute(angle-100)
        r1=d1/20
        r2=d2/20
        flag = 1
        bins = 4
        return bins, flag, d1, d2, r1, r2
    elif 100 < angle < 120:
        d1=np.absolute(angle-100)
        d2=np.absolute(angle-120)
        r1=d1/20
        r2=d2/20
        flag = 1
        bins = 5
        return bins, flag, d1, d2, r1, r2
    elif 120 < angle < 140:
        d1=np.absolute(angle-120)
        d2=np.absolute(angle-140)
        r1=d1/20
        r2=d2/20
        flag = 1
        bins = 6
        return bins, flag, d1, d2, r1, r2
    elif 140 < angle < 160:
        d1=np.absolute(angle-140)
        d2=np.absolute(angle-160)
        r1=d1/20
        r2=d2/20
        flag = 1
        bins = 7
        return bins, flag, d1, d2, r1, r2
    elif 160 < angle <= 180:
        d1 = np.absolute(angle-160)
        d2 = np.absolute(angle-180)
        r1=d1/20
        r2=d2/20
        flag = 1
        bins = 8
        return bins, flag, d1, d2, r1, r2
```

```python
def cell_histogram(grad_mag, grad_angle, cell_size):
    m, n = grad_mag.shape
    M = int(m / cell_size)
    N = int(n / cell_size)
    histograms = np.zeros((M, N, 9))
    for i in range(M):
        for j in range(N):
            for x in range(cell_size):
                for y in range(cell_size):
                    angle = grad_angle[i * cell_size + x][j * cell_size + y]
                    mag = grad_mag[i * cell_size + x][j * cell_size + y]
                    bins, f, distance1, distance2, ratio1, ratio2 = get_bin(angle)
                    if f == 0:
                        histograms[i][j][bins] += mag
                    else:
                        if distance1 >= distance2:
                            histograms[i][j][bins] += min(ratio1,ratio2)*mag
                            if bins + 1 == 9:
                                bins = 0
                            histograms[i][j][bins+1] += max(ratio1,ratio2)*mag
                        else:
                            histograms[i][j][bins] += max(ratio1,ratio2)*mag
                            if bins + 1 == 9:
                                bins = 0
                            histograms[i][j][bins+1] += min(ratio1,ratio2)*mag
                            

    return histograms
```

```python
def normalize_histogram_block(cells_hist):
    e=1e-5
    M, N, bins = cells_hist.shape
    normalized_blocks = np.zeros((M - 1, N - 1, 36))
    for i in range(M - 1):
        for j in range(N - 1):
            unnormalized_blocks = []
            for x in range(2):
                for y in range(2):
                    for z in range(bins):
                        unnormalized_blocks.append(cells_hist[i + x][j + y][z])
            unnormalized_blocks = np.asarray(unnormalized_blocks)
            normalized_block = unnormalized_blocks / np.sqrt(np.sum(np.power(unnormalized_blocks,2)) + e)
            for p in range(bins * 4):
                normalized_blocks[i][j][p] = normalized_block[p]
            return normalized_blocks
            
            
```

```python
def get_names(path):
    names = []
    for root, dirnames, filenames in os.walk(path):
        for filename in filenames:
            _,ext = os.path.splitext(filename)
            if ext in ['.png']:
                names.append(filename)
    return names
```
